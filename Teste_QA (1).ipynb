```python
import pandas as pd
import numpy as np
from datetime import datetime
import sempy.fabric as fabric

# =========================================================
# CONFIGURAÇÕES
# =========================================================
CONFIG = {
    # Regras de distribuição de memória
    "dictionary_attention": 0.80,
    "dictionary_critical": 0.90,
    "data_low_attention": 0.10,
    "data_high_attention": 0.30,
    "data_critical": 0.40,
    "hierarchy_attention": 0.10,
    "hierarchy_critical": 0.20,
    "relationship_attention": 0.05,
    "relationship_critical": 0.10,
    "user_hierarchy_attention": 0.05,
    "user_hierarchy_critical": 0.10,

    # Regras de relacionamento
    "missing_rows_attention": 1,
    "missing_rows_critical": 100,
    "relationship_used_size_pct_attention": 0.15,
    "relationship_used_size_pct_critical": 0.30,
    "relationship_used_size_vs_median_attention": 3,
    "relationship_used_size_vs_median_critical": 5,
    "mm_cardinality_threshold": 500,

    # Regras estruturais
    "small_table_rowcount_threshold": 5000,
    "calc_table_ratio_attention": 0.20,
    "calc_table_ratio_critical": 0.40,
    "single_partition_large_table_attention": 5_000_000,
    "single_partition_large_table_critical": 20_000_000,

    # Regras de incremental refresh
    # "Data Size grande" pode ser avaliado por valor absoluto e/ou relativo
    "large_import_table_data_size_abs": 500_000,   # ajuste conforme seu ambiente
    "large_import_table_data_size_pct_model": 0.10, # 10% do Data Size total do modelo
}

run_date = datetime.now()

relationships_list = []
partitions_list = []
tables_list = []
health_summary_list = []
findings_list = []

# =========================================================
# FUNÇÕES AUXILIARES
# =========================================================
def safe_div(a, b):
    if b in [0, None] or pd.isna(b):
        return 0
    return a / b

def to_numeric_cols(df, cols):
    for col in cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)
    return df

def starts_with_any(value, prefixes):
    value = str(value)
    return any(value.startswith(p) for p in prefixes)

def classify_health(score):
    if score >= 90:
        return "Saudável"
    elif score >= 75:
        return "Atenção"
    elif score >= 50:
        return "Ruim"
    return "Crítico"

def add_finding(
    findings,
    run_date,
    datasetid,
    workspaceid,
    rule_name,
    severity,
    object_type,
    object_name,
    metric_value=None,
    threshold=None,
    details=None
):
    findings.append({
        "run_date": run_date,
        "datasetid": datasetid,
        "workspaceid": workspaceid,
        "rule_name": rule_name,
        "severity": severity,
        "object_type": object_type,
        "object_name": object_name,
        "metric_value": metric_value,
        "threshold": threshold,
        "details": details
    })

def summarize_partition_mode(partitions_df):
    """
    Retorna um dataframe com:
    - Table Name
    - table_mode_resolved
    - partition_count
    """
    if partitions_df.empty:
        return pd.DataFrame(columns=["Table Name", "table_mode_resolved", "partition_count"])

    p = partitions_df.copy()
    p["Mode"] = p["Mode"].astype(str)

    def resolve_mode(series):
        values = sorted(set(series.dropna().astype(str).tolist()))
        if len(values) == 1:
            return values[0]
        elif len(values) == 0:
            return "Unknown"
        return "Mixed"

    mode_summary = (
        p.groupby("Table Name", dropna=False)
         .agg(
             table_mode_resolved=("Mode", resolve_mode),
             partition_count=("Partition Name", "count")
         )
         .reset_index()
    )
    return mode_summary

def is_large_import_table(row, total_data_size_model, config):
    """
    Considera a tabela grande se atender pelo menos um dos critérios:
    - Data Size absoluto alto
    - participação relevante no Data Size total do modelo
    """
    data_size = row.get("Data Size", 0)
    pct_model_data = safe_div(data_size, total_data_size_model)

    return (
        data_size >= config["large_import_table_data_size_abs"]
        or pct_model_data >= config["large_import_table_data_size_pct_model"]
    )

# =========================================================
# INPUT BASE
# =========================================================
values = (
    reports_filtered
    .select("datasetid", "workspaceid")
    .distinct()
    .collect()
)

# =========================================================
# LOOP PRINCIPAL
# =========================================================
for item in values:
    datasetid = item["datasetid"]
    workspaceid = item["workspaceid"]

    try:
        # ---------------------------------------------
        # MEMORY ANALYZER
        # ---------------------------------------------
        memory_analyzer = fabric.model_memory_analyzer(
            dataset=datasetid,
            workspace=workspaceid,
            return_dataframe=True
        )

        relationships_information = memory_analyzer["Relationships"].copy()
        partitions_information = memory_analyzer["Partitions"].copy()
        tables_information = memory_analyzer["Tables"].copy()

        for df in [relationships_information, partitions_information, tables_information]:
            df["datasetid"] = datasetid
            df["workspaceid"] = workspaceid
            df["run_date"] = run_date

        relationships_list.append(relationships_information)
        partitions_list.append(partitions_information)
        tables_list.append(tables_information)

        # ---------------------------------------------
        # TABLES
        # ---------------------------------------------
        t = tables_information.copy()
        t = to_numeric_cols(
            t,
            [
                "Row Count", "Total Size", "Dictionary Size", "Data Size",
                "Hierarchy Size", "Relationship Size", "User Hierarchy Size",
                "Partitions", "Columns", "% DB"
            ]
        )

        t["pct_dictionary"] = t.apply(lambda x: safe_div(x["Dictionary Size"], x["Total Size"]), axis=1)
        t["pct_data"] = t.apply(lambda x: safe_div(x["Data Size"], x["Total Size"]), axis=1)
        t["pct_hierarchy"] = t.apply(lambda x: safe_div(x["Hierarchy Size"], x["Total Size"]), axis=1)
        t["pct_relationship"] = t.apply(lambda x: safe_div(x["Relationship Size"], x["Total Size"]), axis=1)
        t["pct_user_hierarchy"] = t.apply(lambda x: safe_div(x["User Hierarchy Size"], x["Total Size"]), axis=1)

        total_model_size = t["Total Size"].sum()
        total_data_size_model = t["Data Size"].sum()
        table_count = len(t)
        median_table_size = t["Total Size"].median() if len(t) > 0 else 0

        auto_date_count = t["Table Name"].astype(str).str.startswith("LocalDateTable_").sum()
        date_template_count = t["Table Name"].astype(str).str.startswith("DateTableTemplate_").sum()
        calculated_table_count = (t["Type"].astype(str) == "Calculated Table").sum()
        calc_table_ratio = safe_div(calculated_table_count, table_count)

        small_table_high_memory_count = t[
            (t["Row Count"] < CONFIG["small_table_rowcount_threshold"]) &
            (t["Total Size"] > median_table_size) &
            (
                (t["pct_dictionary"] > 0.75) |
                (t["pct_user_hierarchy"] > CONFIG["user_hierarchy_critical"])
            )
        ].shape[0]

        high_dictionary_count = (t["pct_dictionary"] > CONFIG["dictionary_critical"]).sum()
        high_hierarchy_count = (t["pct_hierarchy"] > CONFIG["hierarchy_critical"]).sum()
        high_relationship_size_count = (t["pct_relationship"] > CONFIG["relationship_critical"]).sum()
        high_user_hierarchy_count = (t["pct_user_hierarchy"] > CONFIG["user_hierarchy_critical"]).sum()

        # ---------------------------------------------
        # PARTITIONS / MODE
        # ---------------------------------------------
        p = partitions_information.copy()
        p = to_numeric_cols(p, ["Record Count", "Segment Count", "Records per Segment"])

        partition_mode_summary = summarize_partition_mode(p)

        t = t.merge(
            partition_mode_summary,
            on="Table Name",
            how="left"
        )

        t["table_mode_resolved"] = t["table_mode_resolved"].fillna("Unknown")
        t["partition_count"] = t["partition_count"].fillna(0)

        single_partition_large_tables_attention = t[
            (t["partition_count"] == 1) &
            (t["Row Count"] > CONFIG["single_partition_large_table_attention"])
        ].shape[0]

        single_partition_large_tables_critical = t[
            (t["partition_count"] == 1) &
            (t["Row Count"] > CONFIG["single_partition_large_table_critical"])
        ].shape[0]

        # ---------------------------------------------
        # RELATIONSHIPS
        # ---------------------------------------------
        r = relationships_information.copy()
        if not r.empty:
            r = to_numeric_cols(
                r,
                ["Used Size", "Missing Rows", "Max From Cardinality", "Max To Cardinality"]
            )

            total_relationship_used_size = r["Used Size"].sum()
            median_used_size = r["Used Size"].median() if len(r) > 0 else 0

            r["used_size_pct_of_relationships"] = r["Used Size"].apply(
                lambda x: safe_div(x, total_relationship_used_size)
            )
            r["used_size_vs_median"] = r["Used Size"].apply(
                lambda x: safe_div(x, median_used_size) if median_used_size > 0 else 0
            )
            r["missing_ratio_from"] = r.apply(
                lambda x: safe_div(x["Missing Rows"], x["Max From Cardinality"]),
                axis=1
            )

            mm_relationship_count = (r["Multiplicity"].astype(str) == "m:m").sum()
            missing_rows_relationship_count = (r["Missing Rows"] > 0).sum()
            missing_rows_critical_count = (r["Missing Rows"] > CONFIG["missing_rows_critical"]).sum()

            high_used_size_relationship_count = r[
                (r["used_size_pct_of_relationships"] > CONFIG["relationship_used_size_pct_attention"]) |
                (r["used_size_vs_median"] > CONFIG["relationship_used_size_vs_median_attention"])
            ].shape[0]

            mm_high_used_size_count = r[
                (r["Multiplicity"].astype(str) == "m:m") &
                (
                    (r["used_size_pct_of_relationships"] > CONFIG["relationship_used_size_pct_attention"]) |
                    (r["used_size_vs_median"] > CONFIG["relationship_used_size_vs_median_attention"])
                )
            ].shape[0]

            mm_high_cardinality_count = r[
                (r["Multiplicity"].astype(str) == "m:m") &
                (r["Max From Cardinality"] > CONFIG["mm_cardinality_threshold"]) &
                (r["Max To Cardinality"] > CONFIG["mm_cardinality_threshold"])
            ].shape[0]
        else:
            total_relationship_used_size = 0
            mm_relationship_count = 0
            missing_rows_relationship_count = 0
            missing_rows_critical_count = 0
            high_used_size_relationship_count = 0
            mm_high_used_size_count = 0
            mm_high_cardinality_count = 0

        # ---------------------------------------------
        # ---------------------------------------------
        # INCREMENTAL REFRESH CHECK
        # Validar SOMENTE se:
        # 1) a tabela estiver em modo Import na tabela Partitions
        # 2) a tabela for considerada grande
        # ---------------------------------------------
        incremental_refresh_checked_count = 0
        incremental_refresh_missing_count = 0
        incremental_refresh_error_count = 0
        large_import_table_count = 0

        # merge já feito antes:
        # t = t.merge(partition_mode_summary, on="Table Name", how="left")

        t["table_mode_resolved"] = t["table_mode_resolved"].fillna("Unknown").astype(str)
        t["partition_count"] = t["partition_count"].fillna(0)

        # flag de tabela em Import baseada EXCLUSIVAMENTE em Partitions
        t["is_import_mode_from_partitions"] = (
            t["table_mode_resolved"].str.strip().str.lower() == "import"
        )

        # flag de tabela grande
        t["is_large_table_for_incremental"] = t.apply(
            lambda row: is_large_import_table(row, total_data_size_model, CONFIG),
            axis=1
        )

        # validar incremental refresh SOMENTE para tabelas que satisfazem os 2 critérios
        tables_to_check_incremental = t[
            (t["is_import_mode_from_partitions"] == True) &
            (t["is_large_table_for_incremental"] == True)
        ].copy()

        large_import_table_count = len(tables_to_check_incremental)

        if large_import_table_count > 0:
            try:
                tom = fabric.TOMWrapper(datasetid, workspaceid, readonly=True)
            except Exception as tom_open_error:
                tom = None
                add_finding(
                    findings_list,
                    run_date,
                    datasetid,
                    workspaceid,
                    "TOM_WRAPPER_ERROR",
                    "High",
                    "Dataset",
                    str(datasetid),
                    details=f"Erro ao abrir TOMWrapper: {tom_open_error}"
                )

            if tom is not None:
                for _, row_t in tables_to_check_incremental.iterrows():
                    table_name = row_t["Table Name"]
                    incremental_refresh_checked_count += 1

                    try:
                        has_incremental_refresh = tom.has_incremental_refresh_policy(
                            table_name=table_name
                        )

                        if not has_incremental_refresh:
                            incremental_refresh_missing_count += 1
                            add_finding(
                                findings_list,
                                run_date,
                                datasetid,
                                workspaceid,
                                "LARGE_IMPORT_TABLE_WITHOUT_INCREMENTAL_REFRESH",
                                "High",
                                "Table",
                                table_name,
                                metric_value=row_t["Data Size"],
                                threshold=CONFIG["large_import_table_data_size_abs"],
                                details=(
                                    "Tabela grande em modo Import, identificada pela tabela "
                                    "Partitions, sem política de atualização incremental. "
                                    f"Data Size={row_t['Data Size']}, "
                                    f"pct_data_model={safe_div(row_t['Data Size'], total_data_size_model):.4f}, "
                                    f"partition_mode={row_t['table_mode_resolved']}"
                                )
                            )

                    except Exception as inc_error:
                        incremental_refresh_error_count += 1
                        add_finding(
                            findings_list,
                            run_date,
                            datasetid,
                            workspaceid,
                            "INCREMENTAL_REFRESH_CHECK_ERROR",
                            "Medium",
                            "Table",
                            table_name,
                            metric_value=row_t["Data Size"],
                            threshold=None,
                            details=f"Erro ao validar incremental refresh: {inc_error}"
                        )
        else:
            # nenhuma tabela elegível para validação
            pass

        # ---------------------------------------------
        # SCORE
        # ---------------------------------------------
        health_score = 100

        # Relacionamentos
        health_score -= mm_relationship_count * 15
        health_score -= mm_high_used_size_count * 10
        health_score -= mm_high_cardinality_count * 10
        health_score -= missing_rows_relationship_count * 5
        health_score -= missing_rows_critical_count * 5

        # Estrutura de data automática
        if auto_date_count > 0:
            health_score -= 15
        if date_template_count > 0:
            health_score -= 10

        # Distribuição de memória
        health_score -= high_dictionary_count * 8
        health_score -= high_hierarchy_count * 6
        health_score -= high_relationship_size_count * 6
        health_score -= high_user_hierarchy_count * 8
        health_score -= small_table_high_memory_count * 8

        # Calculated tables
        if calc_table_ratio > CONFIG["calc_table_ratio_critical"]:
            health_score -= 10
        elif calc_table_ratio > CONFIG["calc_table_ratio_attention"]:
            health_score -= 5

        # Partições
        health_score -= single_partition_large_tables_attention * 5
        health_score -= single_partition_large_tables_critical * 5

        # Incremental refresh
        health_score -= incremental_refresh_missing_count * 12
        health_score -= incremental_refresh_error_count * 3

        health_score = max(0, min(100, health_score))
        health_status = classify_health(health_score)

        # ---------------------------------------------
        # SUMMARY
        # ---------------------------------------------
        summary_row = {
            "run_date": run_date,
            "datasetid": datasetid,
            "workspaceid": workspaceid,

            "table_count": table_count,
            "total_model_size": total_model_size,
            "total_data_size_model": total_data_size_model,

            "calculated_table_count": calculated_table_count,
            "calculated_table_ratio": calc_table_ratio,

            "auto_date_count": int(auto_date_count),
            "date_template_count": int(date_template_count),
            "small_table_high_memory_count": int(small_table_high_memory_count),

            "high_dictionary_count": int(high_dictionary_count),
            "high_hierarchy_count": int(high_hierarchy_count),
            "high_relationship_size_count": int(high_relationship_size_count),
            "high_user_hierarchy_count": int(high_user_hierarchy_count),

            "single_partition_large_tables_attention": int(single_partition_large_tables_attention),
            "single_partition_large_tables_critical": int(single_partition_large_tables_critical),

            "mm_relationship_count": int(mm_relationship_count),
            "missing_rows_relationship_count": int(missing_rows_relationship_count),
            "missing_rows_critical_count": int(missing_rows_critical_count),
            "high_used_size_relationship_count": int(high_used_size_relationship_count),
            "mm_high_used_size_count": int(mm_high_used_size_count),
            "mm_high_cardinality_count": int(mm_high_cardinality_count),

             "large_import_table_count": int(large_import_table_count),
            "incremental_refresh_checked_count": int(incremental_refresh_checked_count),
            "incremental_refresh_missing_count": int(incremental_refresh_missing_count),
            "incremental_refresh_error_count": int(incremental_refresh_error_count),

            "health_score": health_score,
            "health_status": health_status
        }

        health_summary_list.append(summary_row)

        # ---------------------------------------------
        # FINDINGS - TABELAS
        # ---------------------------------------------
        for _, row_t in t[t["pct_dictionary"] > CONFIG["dictionary_critical"]].iterrows():
            add_finding(
                findings_list,
                run_date,
                datasetid,
                workspaceid,
                "HIGH_DICTIONARY_SIZE",
                "High",
                "Table",
                row_t["Table Name"],
                metric_value=row_t["pct_dictionary"],
                threshold=CONFIG["dictionary_critical"],
                details="Dictionary Size acima do limite crítico em relação ao Total Size"
            )

        for _, row_t in t[t["pct_hierarchy"] > CONFIG["hierarchy_critical"]].iterrows():
            add_finding(
                findings_list,
                run_date,
                datasetid,
                workspaceid,
                "HIGH_HIERARCHY_SIZE",
                "Medium",
                "Table",
                row_t["Table Name"],
                metric_value=row_t["pct_hierarchy"],
                threshold=CONFIG["hierarchy_critical"],
                details="Hierarchy Size acima do limite crítico em relação ao Total Size"
            )

        for _, row_t in t[t["pct_relationship"] > CONFIG["relationship_critical"]].iterrows():
            add_finding(
                findings_list,
                run_date,
                datasetid,
                workspaceid,
                "HIGH_TABLE_RELATIONSHIP_SIZE",
                "Medium",
                "Table",
                row_t["Table Name"],
                metric_value=row_t["pct_relationship"],
                threshold=CONFIG["relationship_critical"],
                details="Relationship Size da tabela acima do limite crítico"
            )

        for _, row_t in t[t["pct_user_hierarchy"] > CONFIG["user_hierarchy_critical"]].iterrows():
            add_finding(
                findings_list,
                run_date,
                datasetid,
                workspaceid,
                "HIGH_USER_HIERARCHY_SIZE",
                "Medium",
                "Table",
                row_t["Table Name"],
                metric_value=row_t["pct_user_hierarchy"],
                threshold=CONFIG["user_hierarchy_critical"],
                details="User Hierarchy Size acima do limite crítico"
            )

        for _, row_t in t[t["Table Name"].astype(str).str.startswith("LocalDateTable_")].iterrows():
            add_finding(
                findings_list,
                run_date,
                datasetid,
                workspaceid,
                "AUTO_DATE_TABLE",
                "High",
                "Table",
                row_t["Table Name"],
                metric_value=row_t["Total Size"],
                threshold=None,
                details="Tabela automática de data detectada"
            )

        for _, row_t in t[t["Table Name"].astype(str).str.startswith("DateTableTemplate_")].iterrows():
            add_finding(
                findings_list,
                run_date,
                datasetid,
                workspaceid,
                "DATE_TEMPLATE_TABLE",
                "Medium",
                "Table",
                row_t["Table Name"],
                metric_value=row_t["Total Size"],
                threshold=None,
                details="DateTableTemplate detectada no modelo"
            )

        for _, row_t in t[
            (t["Row Count"] < CONFIG["small_table_rowcount_threshold"]) &
            (t["Total Size"] > median_table_size) &
            (
                (t["pct_dictionary"] > 0.75) |
                (t["pct_user_hierarchy"] > CONFIG["user_hierarchy_critical"])
            )
        ].iterrows():
            add_finding(
                findings_list,
                run_date,
                datasetid,
                workspaceid,
                "SMALL_TABLE_HIGH_MEMORY",
                "Medium",
                "Table",
                row_t["Table Name"],
                metric_value=row_t["Total Size"],
                threshold=median_table_size,
                details="Tabela pequena com consumo de memória desproporcional"
            )

        # ---------------------------------------------
        # FINDINGS - RELATIONSHIPS
        # ---------------------------------------------
        if not r.empty:
            for _, row_r in r[r["Multiplicity"].astype(str) == "m:m"].iterrows():
                add_finding(
                    findings_list,
                    run_date,
                    datasetid,
                    workspaceid,
                    "MANY_TO_MANY_RELATIONSHIP",
                    "High",
                    "Relationship",
                    f"{row_r['From Object']} -> {row_r['To Object']}",
                    metric_value=row_r["Used Size"],
                    threshold=None,
                    details="Relacionamento many-to-many detectado"
                )

            for _, row_r in r[r["Missing Rows"] > 0].iterrows():
                severity = "Medium" if row_r["Missing Rows"] <= CONFIG["missing_rows_critical"] else "High"
                add_finding(
                    findings_list,
                    run_date,
                    datasetid,
                    workspaceid,
                    "MISSING_ROWS_RELATIONSHIP",
                    severity,
                    "Relationship",
                    f"{row_r['From Object']} -> {row_r['To Object']}",
                    metric_value=row_r["Missing Rows"],
                    threshold=0,
                    details="Relacionamento com Missing Rows"
                )

            for _, row_r in r[
                (r["used_size_pct_of_relationships"] > CONFIG["relationship_used_size_pct_attention"]) |
                (r["used_size_vs_median"] > CONFIG["relationship_used_size_vs_median_attention"])
            ].iterrows():
                add_finding(
                    findings_list,
                    run_date,
                    datasetid,
                    workspaceid,
                    "HIGH_RELATIONSHIP_USED_SIZE",
                    "Medium",
                    "Relationship",
                    f"{row_r['From Object']} -> {row_r['To Object']}",
                    metric_value=row_r["Used Size"],
                    threshold=CONFIG["relationship_used_size_pct_attention"],
                    details=(
                        f"Used Size relevante. pct_of_relationships={row_r['used_size_pct_of_relationships']:.4f}, "
                        f"vs_median={row_r['used_size_vs_median']:.2f}"
                    )
                )

        # ---------------------------------------------
        # FINDINGS - MODEL LEVEL
        # ---------------------------------------------
        if calc_table_ratio > CONFIG["calc_table_ratio_critical"]:
            add_finding(
                findings_list,
                run_date,
                datasetid,
                workspaceid,
                "HIGH_CALCULATED_TABLE_RATIO",
                "Medium",
                "Dataset",
                str(datasetid),
                metric_value=calc_table_ratio,
                threshold=CONFIG["calc_table_ratio_critical"],
                details="Proporção de calculated tables acima do limite crítico"
            )

    except Exception as e:
        add_finding(
            findings_list,
            run_date,
            datasetid,
            workspaceid,
            "MEMORY_ANALYZER_ERROR",
            "High",
            "Dataset",
            str(datasetid),
            details=str(e)
        )

# =========================================================
# CONSOLIDAÇÃO FINAL
# =========================================================
relationships_history = pd.concat(relationships_list, ignore_index=True) if relationships_list else pd.DataFrame()
partitions_history = pd.concat(partitions_list, ignore_index=True) if partitions_list else pd.DataFrame()
tables_history = pd.concat(tables_list, ignore_index=True) if tables_list else pd.DataFrame()
health_summary = pd.DataFrame(health_summary_list)
findings_df = pd.DataFrame(findings_list)
```

```python
# função para normalizar nomes
def normalize_columns(df):
    new_columns = [
        col.strip().lower().replace(" ", "")
        for col in df.columns
    ]
    
    # renomeia todas as colunas
    df = df.toDF(*new_columns)
    return df
```
